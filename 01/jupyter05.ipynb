##### Dictionary Set
- Dictionary의 키는 반드시 불변(문자열, 숫자 또는 튜플)이어야 한다.
- Dictionary의 키는 유일해야 한다. 

```python
country_codes = {'Finland':'fi', 'South Africa':'za', 'Nepal':'np'}
print(type(country_codes))
print(country_codes.items())
print(country_codes.values())
print(country_codes.keys())
print(len(country_codes))

if country_codes:
    print("country_codes is not empty")
else:
    print("country_codes is empty")
```

```python
days_per_month = {"January":31, "February":28, "March":31}

for month, day in days_per_month.items():
    print(f"{month} is the {day}st")
    
```

```python
roman_numerals = {'Ⅰ':1, 'Ⅱ':2, 'Ⅲ':3, 'Ⅳ':4, 'Ⅴ':5, 'Ⅵ':100}
print(roman_numerals)
print(roman_numerals['Ⅵ'])
roman_numerals['Ⅵ'] = 6
print(roman_numerals)
roman_numerals['Ⅶ'] = 7
print(roman_numerals)
#del roman_numerals['Ⅶ']
roman_numerals.pop('Ⅶ')
print(roman_numerals)
```

```python
print(roman_numerals)
roman_numerals.get('Ⅹ', 'Ⅹ is not in roman_numerals.')
roman_numerals.get('Ⅲ')
'Ⅰ' in roman_numerals
'Ⅷ' in roman_numerals
```

```python
months = {'Jaunary':1, 'February':2, 'March':3}
for month_name in months.keys():
    print(month_name)

for month_number in months.values():
    print(month_number)

months_view = months.keys()

for key in months_view:
    print(key, end=" ")
print()
months['December'] = 12
print(months)
for key in months_view:
    print(key, end=" ")
print()


for month_name in sorted(months.keys()):
    print(month_name, end=",")
```
