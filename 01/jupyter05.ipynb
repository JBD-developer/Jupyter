##### Dictionary Set
- Dictionary의 키는 반드시 불변(문자열, 숫자 또는 튜플)이어야 한다.
- Dictionary의 키는 유일해야 한다. 

```python
country_codes = {'Finland':'fi', 'South Africa':'za', 'Nepal':'np'}
print(type(country_codes))
print(country_codes.items())
print(country_codes.values())
print(country_codes.keys())
print(len(country_codes))

if country_codes:
    print("country_codes is not empty")
else:
    print("country_codes is empty")
```

```python
grade_book = {
    'Susan':[95, 85, 100],
    'Eduardo':[83, 95, 79],
    'Azizi':[91, 89, 82],
    'Pantipa':[97,91, 92]
}

all_grades_total = 0
all_grades_count = 0 

for name, grades in grade_book.items():
    total= sum(grades)
    print(f'The average score of the {name} is {total/len(grades):.2f}')
    all_grades_total += total
    all_grades_count += len(grades)
print(f'Average: { all_grades_total/ all_grades_count:.2f}')
```

```python
text = ('this is sample text with several words this is more sample text with some different words')

word_count = {}

for word in text.split():
    if word in word_count:
        word_count[word]+=1
    else:
        word_count[word]= 1

for key, value in word_count.items():
    print(key, value)
```

```python
from collections import Counter
text = ('this is sample text with several words this is more sample text with some different words')
counter = Counter(text.split())

for word, count in sorted(counter.items()):
    print(f'{word:<12} {count:>10}')
```

```python
print({1, 3, 5} == {1, 5, 3})
print({1, 3, 5} < {3, 5, 1})
print({1, 3, 5} < {4, 3, 5, 1})
print({1, 3}.issubset({3, 4, 1}))
```

```python
set1 = {1, 3, 5}
set2 = {2, 3, 4}
print(set1 | set2)
print(set1.union(set2))
print(set2.union([1, 2, 3, 4, 5, 6, 7, 8]))

print(set1 & set2)
print(set1.intersection(set2))
print("="*30)
print(set1 - set2)
print({1, 3, 5, 7}.difference([2, 2, 3, 3, 4, 4]))

print("="*30)
print(set1 ^ set2)
print({1, 3, 5, 7}.symmetric_difference([2, 2, 3, 3, 4, 4]))

```
