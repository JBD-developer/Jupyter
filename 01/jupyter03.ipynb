```python
def maximum(value1, value2, value3):
    """Returns the largest number of three values"""
    max_value = value1
    if max_value < value2:
        max_value = value2
    if max_value < value3:
        max_value = value3
    
    return max_value

print(maximum(12, 28, 36))
print(maximum(12.3, 45.6, 9.7))
print(maximum('yellow', 'red', 'orange'))

maximum??
```

```python
import random

def roll_dice():
    """Throw two dice and return the number of each die with a tuple"""
    die1 = random.randrange(1, 7)
    die2 = random.randrange(1, 7)
    return die1, die2 

def display_dice(dice):
    """Output two dice numbers at once"""
    die1, die2 = dice 
    print(f"The dice thrown by the player {die1} + {die2} = {sum(dice)}")

die_value = roll_dice() # first game 
display_dice(die_value)
sum_of_dice = sum(die_value)

if sum_of_dice in (7, 11):
    game_status = "Win"

elif sum_of_dice in (2, 3, 12): 
    game_status = "Lose"

else :
    game_status = "Continue"
    my_point = sum_of_dice
    print("Point :", my_point)
    
while game_status == "Continue": 
    die_value = roll_dice()
    display_dice(die_value)
    sum_of_dice = sum(die_value)
    
    if sum_of_dice == my_point:
        game_status = "Win"
    elif sum_of_dice == 7:
        game_status = "Lose"

if game_status == "Win":
    print("Player Win")
else:
    print("Player Lose")    
     
```
