```python
import random

def roll_dice():
    """Throw two dice and return the number of each die with a tuple"""
    die1 = random.randrange(1, 7)
    die2 = random.randrange(1, 7)
    return die1, die2 

def display_dice(dice):
    """Output two dice numbers at once"""
    die1, die2 = dice 
    print(f"The dice thrown by the player {die1} + {die2} = {sum(dice)}")

die_value = roll_dice() # first game 
display_dice(die_value)
sum_of_dice = sum(die_value)

if sum_of_dice in (7, 11):
    game_status = "Win"

elif sum_of_dice in (2, 3, 12): 
    game_status = "Lose"

else :
    game_status = "Continue"
    my_point = sum_of_dice
    print("Point :", my_point)
    
while game_status == "Continue": 
    die_value = roll_dice()
    display_dice(die_value)
    sum_of_dice = sum(die_value)
    
    if sum_of_dice == my_point:
        game_status = "Win"
    elif sum_of_dice == 7:
        game_status = "Lose"

if game_status == "Win":
    print("Player Win")
else:
    print("Player Lose")    
     
```

# identity
- id를 호출해서 얻는 정수값을 객체의 식별값(identity)라고 한다. 메모리에 있는 서로 다른 두 객체는 동일한 식별값을 가질 수 없다.
- 객체의 식별값을 통해서 객체가 참조로 전달되었는지를 확인하는데 사용한다. 

```python
def cube(number):
    print('id(number) :', id(number))    
    return number ** 3 

x = 7 
print(id(x))
print(cube(7))
print(x)

```

```python
def factorial(number):
    """calculation Factorial"""
    if number <= 1:
        return 1
    return number * factorial(number -1 )

for i in range(11):
    print(f"{i}! = {factorial(i)}")
```
