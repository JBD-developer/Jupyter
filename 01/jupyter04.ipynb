```python
# lazy evaluation greedy evalution 

numbers = [10, 3, 7, 1, 9, 4, 2, 8, 5, 6]

for value in(x ** 2 for x in numbers if x % 2 != 0):
    print(value, end=',')

squares_of_odds = (x ** 2 for x in numbers if x % 2 != 0)
print(squares_of_odds)
```

```python
# filter() 
numbers = [10, 3, 7, 1, 9, 4, 2, 8, 5, 6]

def is_odd(x):
    return x % 2 != 0

odd_list = list(filter(is_odd, numbers))
print(odd_list)

odd_list2 = list(filter(lambda x : x % 2 !=0, numbers))
print(odd_list2)
```

```python
# map()
numbers 
print(numbers)

list_map = list(map(lambda x : x ** 2, numbers))
print(list_map)

list_result = list(map(lambda x : x ** 2, filter(lambda x : x % 2 !=0, numbers)))
print(list_result)
```

```python
# zip

names = ["Bob", "Sue", "Amanda"]
grade_point_averages = [3.5, 4.0, 3.75]
for name, gpa in zip(names, grade_point_averages):
    print(f'Name={name}; GPA:{gpa}')
```

```python
a =[[77, 68, 86, 73], 
    [96, 87, 89, 81], 
    [70, 90, 86, 81]]

for row in a:
    for item in row:
        print(item, end=' ')
    print()

for i, row in enumerate(a):
    for j, item in enumerate(row):
        print(f'a[{i}][{j}]= {item}', end= ' ')
    print()
```
