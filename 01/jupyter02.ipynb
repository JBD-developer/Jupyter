```python
total = 0 
grade_counter = 0
grades = [98, 76, 71, 87, 83, 90, 57, 79, 82, 94]

for i in grades:
    total += i 
    grade_counter +=1 

average = total/grade_counter
print("average : %f" % average)
```

```python
# sentinel 

total = 0
grade_counter = 0 

grade = int(input("If there's a score, if there's no score, enter -1."))

while grade != -1:
    total += grade 
    grade_counter +=1
    grade = int(input("If there's a score, if there's no score, enter -1."))

if grade_counter != 0:
    average = total / grade_counter
    print("Average : %f" %average)
else :
    print("There is no score entered")
    
```

```python
from decimal import Decimal

principal = Decimal('1000.00')
print(type(principal), principal)
rate = Decimal('0.05')

for year in range(1, 11):
    amount = principal * (1 + rate) ** year
    print(f"{year:>2}{amount:>10.2f}")
```
