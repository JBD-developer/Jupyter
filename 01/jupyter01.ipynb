```python
7 >= 3
7 <= 4 

number1 = int(input("Enter the first integer"))
number2 = int(input("Enter the second integer"))

if number1 == number2:
    print(number1, "and", number2 , "are the same")

if number1 != number2:
    print(number1, "and", number2 , "are not the same")

if number1 < number2:
    print(number1, "is less than" ,number2)

if number1 > number2:
    print(number1, "is bigger than" ,number2)

if number1 <= number2:
    print(number1, "is less than or equal to" ,number2)

if number1 >= number2:
    print(number1, "is bigger than or equal to" ,number2)
```

```python
num1 = int(input("Enter the first number"))
num2 = int(input("Enter the second number"))
num3 = int(input("Enter the third number"))

mininum = num1 

if num2 < mininum:
    mininum = num2
if num3 < mininum:
    mininum = num3

print("Minimum value %d" %mininum)

lst = []
lst.append(num1)
lst.append(num2)
lst.append(num3)

# reduction 
print("Max value : %d" %max(lst))
print("Min value : %d" %min(lst))
```
